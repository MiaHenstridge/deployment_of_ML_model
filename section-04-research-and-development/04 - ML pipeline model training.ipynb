```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# to save the model
import joblib

# to build the model
from sklearn.linear_model import Lasso

# to evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

# to visualize all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
# Load the train and test set with the engineered variables

X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

X_train.head()
```

```python
# Load the target (log-transformed)
y_train = pd.read_csv('ytrain.csv')
y_test = pd.read_csv('ytest.csv')

y_train.head()
```

```python
# Load the preselected features

features = pd.read_csv('selected_features.csv')
features = features['0'].to_list()
```

```python
features
```

```python
# grab the selected features from train and test set
X_train = X_train[features]
X_test = X_test[features]
```

## Regularized linear regression: Lasso

```python
# set up the model
# with the seed

lin_model = Lasso(alpha=0.001, random_state=0)

# train the model
lin_model.fit(X_train, y_train)
```

```python
# evaluated the model

# remember that we log transformed the output (SalePrice)
# In order to get the true performance of the Lasso
# we need to transform both the target and the predictions
# back to the original house price values

# We will evaluate performance using the MSE and RMSE and R2

# make predictions for the train set
pred = lin_model.predict(X_train)
```

```python
# determine mse, rmse, and r2 for train set
print(f'train mse: {int(mean_squared_error(np.exp(y_train), np.exp(pred)))}')
print(f'train rmse: {int(mean_squared_error(np.exp(y_train), np.exp(pred), squared=False))}')
print(f'train r2: {r2_score(np.exp(y_train), np.exp(pred))}')
```

```python
# make predictions for test set
pred = lin_model.predict(X_test)

# determine mse, rmse, and r2 for test set
print(f'test mse: {int(mean_squared_error(np.exp(y_test), np.exp(pred)))}')
print(f'test rmse: {int(mean_squared_error(np.exp(y_test), np.exp(pred), squared=False))}')
print(f'test r2: {r2_score(np.exp(y_test), np.exp(pred))}')
```

```python
print(f'Average house price: {int(np.exp(y_train).median())}')
```

```python
# evaluate predictions with respect to the real sale price
plt.scatter(y_test, lin_model.predict(X_test))
plt.xlabel('True House Price')
plt.ylabel('Predicted House Price')
plt.title('Evaluation of Lasso Predictions')
```

```python
y_test.reset_index(drop=True)
```

```python
# evaluate the distribution of errors
# they should be fairly normally distributed

y_test.reset_index(drop=True, inplace=True)

preds = pd.Series(lin_model.predict(X_test))

preds
```

```python
errors = y_test['SalePrice'] - preds
errors.hist(bins=30)
plt.show()
```

## Feature importance

```python
importance = pd.Series(np.abs(lin_model.coef_.ravel()))
importance.index = features
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
plt.ylabel('Lasso Coefficients')
plt.xlabel('Feature Importance')
```

## Save the model

```python
joblib.dump(lin_model, 'linear_regression.joblib')
```
